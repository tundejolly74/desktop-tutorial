## Pre-processing the Data

Changing the type of the datasets from a string value to an integer. This is so we can use the type as a label.

```python
white['type'] = 0
red['type']   = 1
```

```python
white.head()
```

We will connect the two datasets to be one, to feed into the ANN.

```python
wines = red.append(white, ignore_index=True)

wines
```

Creating values for the data (X) and labels y.

```python
X = wines
y = np.ravel(wines.type)
```

```python
print(y)
```

```python
# Dropping the targets from the dataframe
X = X.drop(columns=['type'])
```

## Creating training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

### Standardising the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler().fit(X_train)
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)
```

**Creating the ANN**

```python
import tensorflow as tf

from tensorflow              import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

creating the model:

```python
model = Sequential()
```

creating the model's layers:

```python
# Input layer
model.add(Dense(12, activation='relu', input_shape=(12,)))

#Hidden layer
model.add(Dense(8, activation='relu'))

# Output label
model.add(Dense(1, activation='sigmoid'))
```

Exploring the models parameters:

```python
print("Output shape:", model.output_shape)
```

```python
print("Model weights:", model.get_weights())
```

Compile the model:

```python
model.compile(loss='binary_crossentropy', optimizer='adam',
              metrics=['accuracy'])
```

Fit the model:

```python
model.fit(X_train, y_train, epochs=20,
          batch_size=1, verbose=1)
```

Predict with the model:


```python
y_pred = model.predict(X_test)
y_pred[:15]
```

```python
predictions = [1 if p > 0.5 else 0 for p in y_pred]
```

```python
predictions[:15]
```

```python
y_test[:15]
```

Evaluating the model:

```python
score = model.evaluate(X_test, y_test, verbose=1)
score
```

```python
from sklearn.metrics import confusion_matrix, precision_score
```

```python
confusion_matrix(y_test, predictions)
```

```python
precision_score(y_test, y_pred.round())
```
